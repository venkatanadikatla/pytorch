```python
# Load necessary library
library(coda)

# Death data for the five causes from 2018 to 2022
deaths <- matrix(c(33058, 34452, 31315, 24516, 24614,   # Accidents
                   6712, 7571, 31315, 5341, 5234,       # Homicide
                   8663, 8862, 8454, 8059, 8020,       # Suicide
                   3789, 4155, 3984, 3495, 3561,       # Heart Disease
                   3641, 3615, 3573, 3577, 3684),      # Malignant Neoplasms
                 ncol = 5, byrow = TRUE)

n_years <- nrow(deaths)
n_causes <- ncol(deaths)

# Function to implement the separate Gaussian model using Gibbs sampling
separate_gaussian <- function(deaths, n_iter = 5000, burn_in = 1000) {
  n_years <- nrow(deaths)
  n_causes <- ncol(deaths)

  # Prior hyperparameters for mu and sigma^2
  mu_prior_mean <- 0
  mu_prior_variance <- 100^2  # Non-informative prior
  alpha <- 0.001  # Weakly informative prior for sigma^2
  beta <- 0.001

  # Initialize parameters
  mu <- apply(deaths, 2, mean)  # Starting values for mu
  sigma2 <- 1000  # Initial value for variance

  # Storage for MCMC samples
  mu_samples <- matrix(NA, nrow = n_iter, ncol = n_causes)
  sigma2_samples <- numeric(n_iter)

  # Gibbs Sampling
  for (iter in 1:n_iter) {
    # Update mu for each cause
    for (c in 1:n_causes) {
      y_c <- deaths[, c]
      n <- length(y_c)
      # Posterior mean and variance for mu
      mu_post_mean <- (sum(y_c) / sigma2 + mu_prior_mean / mu_prior_variance) /
                      (n / sigma2 + 1 / mu_prior_variance)
      mu_post_var <- 1 / (n / sigma2 + 1 / mu_prior_variance)
      mu[c] <- rnorm(1, mean = mu_post_mean, sd = sqrt(mu_post_var))
    }

    # Update sigma^2
    residuals <- deaths - matrix(rep(mu, each = n_years), n_years, n_causes)
    ss <- sum(residuals^2)
    sigma2_post_shape <- alpha + 0.5 * n_years * n_causes
    sigma2_post_rate <- beta + 0.5 * ss
    sigma2 <- 1 / rgamma(1, shape = sigma2_post_shape, rate = sigma2_post_rate)

    # Store samples
    mu_samples[iter, ] <- mu
    sigma2_samples[iter] <- sigma2
  }

  # Remove burn-in period
  mu_samples <- mu_samples[(burn_in+1):n_iter, ]
  sigma2_samples <- sigma2_samples[(burn_in+1):n_iter]

  list(mu_samples = mu_samples, sigma2_samples = sigma2_samples)
}

# Function to run multiple chains of the separate Gaussian model
separate_gaussian_multiple_chains <- function(deaths, n_iter = 5000, burn_in = 1000, n_chains = 2) {
  chains <- list()

  for (chain in 1:n_chains) {
    chains[[chain]] <- separate_gaussian(deaths, n_iter, burn_in)
  }

  chains
}

# Running two chains
n_chains <- 2
chains_results <- separate_gaussian_multiple_chains(deaths, n_iter = 5000, burn_in = 1000, n_chains = n_chains)

# Extract the mu_samples from each chain
mu_samples_chain1 <- chains_results[[1]]$mu_samples
mu_samples_chain2 <- chains_results[[2]]$mu_samples

# Convert samples to mcmc objects
chain1_mcmc <- as.mcmc(mu_samples_chain1)
chain2_mcmc <- as.mcmc(mu_samples_chain2)

# Combine chains into an mcmc.list
mcmc_list <- as.mcmc.list(list(chain1_mcmc, chain2_mcmc))

# Gelman-Rubin diagnostic for convergence
gelman_diag <- gelman.diag(mcmc_list)

# Display Gelman-Rubin diagnostic
print(gelman_diag)

# Optional: Trace plots for mu_samples to assess convergence
traceplot(mcmc_list)

# Posterior means of mu for each cause
posterior_means_separate <- apply(mu_samples_chain1, 2, mean)
print(posterior_means_separate)

# Predictive mean for a 6th hypothetical cause
mu_6_separate <- rnorm(1, mean = mean(posterior_means_separate),
                       sd = sqrt(mean(chains_results[[1]]$sigma2_samples)))

# Predictive distribution for the 6th cause (y_pred)
y_pred_separate <- rnorm(1000, mean = mu_6_separate,
                         sd = sqrt(mean(chains_results[[1]]$sigma2_samples)))
print(y_pred_separate)

# Display the predictive mean for the 6th cause
print(mu_6_separate)

# Visualize the predictive distribution for the 6th cause
hist(y_pred_separate, main = "Predictive Distribution for 6th Cause",
     xlab = "Deaths", breaks = 30, col = "lightblue")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, gamma

# Death data for the five causes from 2018 to 2022
deaths = np.array([[33058, 34452, 31315, 24516, 24614],   # Accidents
                   [6712, 7571, 31315, 5341, 5234],       # Homicide
                   [8663, 8862, 8454, 8059, 8020],        # Suicide
                   [3789, 4155, 3984, 3495, 3561],        # Heart Disease
                   [3641, 3615, 3573, 3577, 3684]])       # Malignant Neoplasms

n_years, n_causes = deaths.shape

# Function to implement the separate Gaussian model using Gibbs sampling
def separate_gaussian(deaths, n_iter=5000, burn_in=1000):
    n_years, n_causes = deaths.shape

    # Prior hyperparameters for mu and sigma^2
    mu_prior_mean = 0
    mu_prior_variance = 100**2  # Non-informative prior
    alpha = 0.001  # Weakly informative prior for sigma^2
    beta = 0.001

    # Initialize parameters
    mu = np.mean(deaths, axis=0)  # Starting values for mu (mean for each cause)
    sigma2 = 1000  # Initial value for variance

    # Storage for MCMC samples
    mu_samples = np.zeros((n_iter, n_causes))
    sigma2_samples = np.zeros(n_iter)

    # Gibbs Sampling
    for iter in range(n_iter):
        # Update mu for each cause
        for c in range(n_causes):
            y_c = deaths[:, c]
            n = len(y_c)
            # Posterior mean and variance for mu
            mu_post_mean = (np.sum(y_c) / sigma2 + mu_prior_mean / mu_prior_variance) / \
                           (n / sigma2 + 1 / mu_prior_variance)
            mu_post_var = 1 / (n / sigma2 + 1 / mu_prior_variance)
            mu[c] = np.random.normal(mu_post_mean, np.sqrt(mu_post_var))

        # Update sigma^2
        residuals = deaths - mu
        ss = np.sum(residuals**2)
        sigma2_post_shape = alpha + 0.5 * n_years * n_causes
        sigma2_post_rate = beta + 0.5 * ss
        sigma2 = 1 / np.random.gamma(sigma2_post_shape, 1 / sigma2_post_rate)

        # Store samples
        mu_samples[iter, :] = mu
        sigma2_samples[iter] = sigma2

    # Remove burn-in period
    mu_samples = mu_samples[burn_in:]
    sigma2_samples = sigma2_samples[burn_in:]

    return mu_samples, sigma2_samples

# Function to run multiple chains of the separate Gaussian model
def separate_gaussian_multiple_chains(deaths, n_iter=5000, burn_in=1000, n_chains=2):
    chains = []

    for chain in range(n_chains):
        chains.append(separate_gaussian(deaths, n_iter, burn_in))

    return chains

# Running two chains
n_chains = 2
chains_results = separate_gaussian_multiple_chains(deaths, n_iter=5000, burn_in=1000, n_chains=n_chains)

# Extract the mu_samples from each chain
mu_samples_chain1 = chains_results[0][0]
mu_samples_chain2 = chains_results[1][0]

# Posterior means of mu for each cause (from the first chain)
posterior_means_separate = np.mean(mu_samples_chain1, axis=0)
print("Posterior means (separate model) for each cause:")
print(posterior_means_separate)

# Predictive mean for a 6th hypothetical cause
mu_6_separate = np.random.normal(np.mean(posterior_means_separate), np.sqrt(np.mean(chains_results[0][1])))
print(f"Predictive mean for hypothetical 6th cause: {mu_6_separate}")

# Predictive distribution for the 6th cause (y_pred)
y_pred_separate = np.random.normal(mu_6_separate, np.sqrt(np.mean(chains_results[0][1])), 1000)

# Display the predictive mean for the 6th cause
print(f"Predictive distribution for 6th cause (sample of 5): {y_pred_separate[:5]}")

# Visualize the predictive distribution for the 6th cause
plt.hist(y_pred_separate, bins=30, color="lightblue")
plt.title("Predictive Distribution for 6th Cause")
plt.xlabel("Deaths")
plt.show()
```
